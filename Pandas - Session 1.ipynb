## Drop Duplicates

    subset - based on which columns the data duplication to be checked
    keep - First is the by default parameter which is applied to check the first values.


```python

# Drop duplicate values bases on single columns 

data.drop_duplicates(subset=['city'])
```

```python
# Drop duplicate values bases on 2 columns 

# It is working as AND operation

data.drop_duplicates(subset=['city','season']).shape
```

```python

# Using drop_duplicates which we can check on which year which team was the winner

(data.drop_duplicates(subset=['season'],keep='last')[['season','winner']]).sort_values('season')
```
